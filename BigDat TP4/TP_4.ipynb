###Importar

```python
# Importar bibliotecas
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Carga de datos
data = pd.read_csv('/content/drive/MyDrive/Big Data/TP4/Crimenes.csv')
```

###Calculos "cruzados"

```python
# Seleccionar las características relevantes
features = data[['Asesinatos', 'Asaltos', 'Poblacion Urbana', 'Violaciones']]

# Normalizar los datos
features_normalized = (features - features.mean()) / features.std()

# Aplicar K-Means con k=2 (dos clústeres para zonas de baja y alta peligrosidad)
kmeans = KMeans(n_clusters=2, random_state=42)
data['Cluster'] = kmeans.fit_predict(features_normalized)

# Visualizar los resultados con nombres de estados
plt.figure(figsize=(10, 6))
plt.scatter(features_normalized['Asesinatos'], features_normalized['Asaltos'], c=data['Cluster'], cmap='viridis')

# Mostrar nombres de estados en los puntos
for i, state in enumerate(data['Estado']):
    plt.text(features_normalized['Asesinatos'][i], features_normalized['Asaltos'][i], state, fontsize=8)

plt.title('Clustering - Asaltos finalizados en Asesinatos')
plt.xlabel('Asesinatos')
plt.ylabel('Asaltos')
plt.show()
```
